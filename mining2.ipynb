# Goal 

The goal of collecting the dataset for "Google stock prediction" is to develop a predictive models that can forecast future closing price of a stock in a trading day, as the datasets provide an excellent opportunity to develop models that can learn patterns and trends, enabling them to make predictions on future new data , aiding investors and analysts in making informed decisions about buying , selling ,or holding google stock






# Dataset Link
https://www.kaggle.com/datasets/shreenidhihipparagi/google-stock-prediction
